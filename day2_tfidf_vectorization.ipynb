What is TF-IDF?

The full name is Term Frequency - Inverse Document Frequency. Sounds scary, but the idea is simple.
Imagine you're comparing anime by their genres. Two things matter:
TF (Term Frequency): How often does a genre appear in THIS anime's description? If "Action" appears, it gets a score.
IDF (Inverse Document Frequency): How rare is this genre across ALL anime? If "Action" appears in 90% of all anime — it's not very useful for telling anime apart. But if "Psychological" only appears in 5% of anime — it's a very distinctive, meaningful genre.
